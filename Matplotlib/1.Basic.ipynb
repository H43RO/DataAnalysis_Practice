```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, np.pi * 10, 500)  # Pi * 10 너비 (31.4 너비)에, 500개의 점을 균일하게 찍음
fig, axes = plt.subplots(2, 1)  # 2개의 그래프가 들어가는 Figure 생성 (2행 1열 형태)
axes[0].plot(x, np.sin(x))  # 첫 번째 그래프는 사인 그래프
axes[1].plot(x, np.cos(x))  # 두 번째 그래프는 코사인 그래프
fig.savefig("sin&cos.png")
```

```python
# 선 그래프 그리기

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-9, 10)
y = x ** 2  # x 제곱 형태의 그래프 (-9부터 +9 까지 범위)
# 라인 스타일로는 '-', ':', '-.', '--' 등을 사용할 수 있음
plt.plot(x, y, linestyle="--", marker="*")
# X축 및 Y축 에서 특정 범위를 자를 수도 있음
plt.show()
```

```python
# 선 그래프 그리기

import matplotlib.pyplot as plt
import numpy as np

x =np.arange(-9, 10)
y1 = x ** 2
y2 = -x
plt.plot(x, y1, linestyle="--", marker="*", color="red", label="y = x * x")
plt.plot(x, y2, linestyle=":", marker="o", color="blue", label="y = -x")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(  # label 의 정보를 알려주는 역할
    shadow=True,
    borderpad=1
)
plt.show()
```
