```python
import matplotlib.pyplot as plt

plt.title("Plot")
plt.plot([1, 4, 9, 16])
plt.show()

plt.title("X ticks")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.show()
```
