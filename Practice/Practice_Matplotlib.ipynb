```python
# 그리드 사용하기

import matplotlib.pyplot as plt

plt.title("Plot")
plt.plot([1, 4, 9, 16])
plt.show()

plt.title("X ticks")
plt.plot([10, 20, 30, 40], [1, 4, 9, 16])
plt.grid(True)
plt.show()
```

```python
# 스타일 사용하기

import matplotlib.pyplot as plt

plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b", lw=5, ls="--",
         marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.grid(True)
plt.show()
```

```python
# 그래프 범위 지정

import matplotlib.pyplot as plt

plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c="b", lw=5, ls="--",
         marker="o", ms=15, mec="g", mew=5, mfc="r")
plt.xlim(0, 50)
plt.ylim(-10, 30)
plt.grid(True)
plt.show()
```

```python
# Tick 설정해주기

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi, np.pi, 256)
c = np.cos(x)
plt.plot(x, c)
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, +1])
plt.grid(True)
plt.show()
```
